For this notebook, we try out a few different classification techniques:
- k-nearest neighbors
- fisher LDA
- LDA
- QDA
- small neural net

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tqdm
from sklearn.model_selection import train_test_split
from sklearn import datasets, svm, metrics

# load data
X = np.load("data/data_X.npy")
bad_y = np.load("data/data_y.npy")
X_dft = np.load("data/X_dft.npy")
X_dwt = np.load("data/X_dwt.npy")

y = bad_y
for i in range(y.shape[0]):
    if bad_y[i] > 5:
        y[i] = bad_y[i] - 1
        
print(np.unique(y, return_counts=True))

# split into train and test (80-20 split)
N_train = int(0.8 * X.shape[0])
N_test = int(0.2 * X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_dft_train, X_dft_test, _, _ = train_test_split(X_dft, y, test_size=0.2)
X_dwt_train, X_dwt_test, _, _ = train_test_split(X_dwt, y, test_size=0.2)
```

### K-nearest neighbors

```python
# First let's write a function to do k nearest neighbors
from sklearn.neighbors import KNeighborsClassifier

def run_knn(X, y, n_neighbors=3):
    knn_clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn_clf.fit(X, y)
    
    return knn_clf

def predict_knn(X, y, clf):
    y_pred = clf.predict(X)
    acc = np.sum(y_pred == y) / y.shape[0]
    print("Accuracy is: ", acc)
    
    return y_pred

# raw data
knn_clf = run_knn(X_train, y_train, 1)
y_pred_test = predict_knn(X_test, y_test, knn_clf)
# dft
knn_clf_dft = run_knn(X_dft_train, y_train, 2)
y_dft_pred_test = predict_knn(X_dft_test, y_test, knn_clf_dft)
# dwt
knn_clf_dwt = run_knn(X_dwt_train, y_train, 2)
y_dwt_pred_test = predict_knn(X_dwt_test, y_test, knn_clf_dwt)
```

### Linear Discriminant Analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def run_lda(X, y):
    lda_clf = LinearDiscriminantAnalysis()
    lda_clf.fit(X, y)
    
    return lda_clf

def predict_lda(X, y, clf):
    y_pred = clf.predict(X)
    acc = np.sum(y_pred == y) / y.shape[0]
    print("Accuracy is: ", acc)
    
    return y_pred

# raw data
lda_clf = run_lda(X_train, y_train)
y_pred_test = predict_lda(X_test, y_test, lda_clf)
# dft
lda_clf_dft = run_lda(X_dft_train, y_train)
y_dft_pred_test = predict_lda(X_dft_test, y_test, lda_clf_dft)
# dwt
lda_clf_dwt = run_knn(X_dwt_train, y_train)
y_dwt_pred_test = predict_knn(X_dwt_test, y_test, lda_clf_dwt)
```

### Quadratic Discriminant Analysis

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

def run_qda(X, y):
    qda_clf = QuadraticDiscriminantAnalysis()
    qda_clf.fit(X, y)
    
    return qda_clf

def predict_qda(X, y, clf):
    y_pred = clf.predict(X)
    acc = np.sum(y_pred == y) / y.shape[0]
    print("Accuracy is: ", acc)
    
    return y_pred

qda_clf = run_qda(X_train, y_train)
predict_qda(X_test, y_test, qda_clf)

qda_clf_dft = run_qda(X_dft_train, y_train)
predict_qda(X_dft_test, y_test, qda_clf_dft)

qda_clf_dwt = run_qda(X_dwt_train, y_train)
predict_qda(X_dwt_test, y_test, qda_clf_dwt)
```

### Small neural net

```python
import random
import torch
import torch.nn as nn
from torch import Tensor
from torch.utils.data import DataLoader, Dataset
import torch.optim as optim
import torch.nn.functional as F

# some constants
dim = X_train.shape[1]
num_train = X_train.shape[0]
num_test = X_test.shape[0]

# first let's scale the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

X_train_reshape = np.reshape(X_train_scaled, (num_train, 1, dim))
X_test_reshape = np.reshape(X_test_scaled, (num_test, 1, dim))

classes, class_pop = np.unique(y_train, return_counts=True)

class_weights = torch.from_numpy(1/class_pop)
print(class_weights)
```

```python
# custom dataset for multiclass
class ClassifierDataset(Dataset):
    
    def __init__(self, X_data, y_data):
        self.X_data = X_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.X_data)

train_dataset = ClassifierDataset(torch.from_numpy(X_train_reshape).float(), torch.from_numpy(y_train).long())
test_dataset = ClassifierDataset(torch.from_numpy(X_test_reshape).float(), torch.from_numpy(y_test).long())
batch_size = 64
train_load = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# calculates accuracy of the model on the dataset
def get_multiclass_accuracy(model, data):
    loader = torch.utils.data.DataLoader(data, batch_size=100)
    correct, total = 0, 0
    for xs, ts in loader:
        zs = model(xs)
        y_pred_softmax = torch.log_softmax(zs, dim = 1)
        _, y_pred_tags = torch.max(y_pred_softmax, dim = 1)
        correct_pred = (ts==y_pred_tags).float()
        correct += torch.sum(correct_pred).item()
        total += int(ts.shape[0])
    return correct / total
```

```python
class multiclass_nn(nn.Module):
    def __init__(self, num_feature, num_class):
        super(multiclass_nn, self).__init__()
        
        self.conv_1 = nn.Conv1d(in_channels=1, out_channels=3, kernel_size=2, stride=2)
        self.conv_2 = nn.Conv1d(in_channels=3, out_channels=2, kernel_size=2, stride=2)
        
        self.layer_1 = nn.Linear(2*24, 512)
        self.layer_2 = nn.Linear(512, 128)
        self.layer_3 = nn.Linear(128, 64)
        self.layer_out = nn.Linear(64, num_class) 
        
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.2)
        self.batchnorm1 = nn.BatchNorm1d(512)
        self.batchnorm2 = nn.BatchNorm1d(128)
        self.batchnorm3 = nn.BatchNorm1d(64)
        
    def forward(self, x):
        
        x = self.conv_1(x)
        x = self.relu(x)
        x = self.conv_2(x)
        x = self.relu(x)
        x = x.view(-1, 2*24)
        
        x = self.layer_1(x)
        x = self.batchnorm1(x)
        x = self.relu(x)
        
        x = self.layer_2(x)
        x = self.batchnorm2(x)
        x = self.relu(x)
        x = self.dropout(x)
        
        x = self.layer_3(x)
        x = self.batchnorm3(x)
        x = self.relu(x)
        x = self.dropout(x)
        
        x = self.layer_out(x)
        
        return x
```

```python
# training
model = multiclass_nn(num_feature = dim, num_class = 22)
num_epochs = 200
learning_rate = 1e-3
lambda_reg = 0.01
iters, losses = [], []
iters_sub, train_acc, test_acc = [], [] ,[]

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss(weight=class_weights.float())
itercount = 0 # the number of iterations

for epoch in range(num_epochs):
    for sample, label in iter(train_load):
        prediction = model(sample)
        weights = torch.cat([x.view(-1) for x in model.parameters()])
        loss = criterion(prediction, label.long())
        
        # add regularization
        l2_reg = None
        for W in model.parameters():
            if l2_reg is None:
                l2_reg = W.norm(2)
            else:
                l2_reg = l2_reg + W.norm(2)
        loss = loss + lambda_reg * l2_reg
        
        loss.backward() # compute updates for each parameter
        optimizer.step() # make the updates for each parameter
        optimizer.zero_grad() # a clean up step for PyTorch
        # save the current training information
        iters.append(itercount)
        losses.append(float(loss)/batch_size) # compute *average* loss
        if itercount % 10 == 0:
            iters_sub.append(itercount)
            train_acc.append(get_multiclass_accuracy(model, train_dataset))
            test_acc.append(get_multiclass_accuracy(model, test_dataset))
        if itercount % 100 == 0:
            print("-"*10, "Iteration:", itercount, "-"*10)
            print("Loss:", float(loss)/batch_size)
            print("Training accuracy:", get_multiclass_accuracy(model, train_dataset))
            print("Validation accuracy:", get_multiclass_accuracy(model, test_dataset))
            print("="*50)
        # increment the iteration number
        itercount += 1
```

```python
# plotting
plt.title("Training Curve")
plt.plot(iters, losses, label="Train")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
plt.title("Accuracy Curve")
plt.plot(iters_sub[:1270], train_acc, label="Train")
plt.plot(iters_sub[:1270], test_acc, label="Test")
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.legend(loc='best')
plt.show()

print("Training accuracy is: ", get_multiclass_accuracy(model, train_dataset))
print("Test accuracy is: ", get_multiclass_accuracy(model, test_dataset))
```
